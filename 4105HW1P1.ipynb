```python
file_path = '/content/drive/My Drive/Machine Learning/Datasets/D3.csv'
df = pd.DataFrame(pd.read_csv(file_path))
df.head()
# checking to make sure dataset access works
```

```python
x1 = df.values[:, 0]
x2 = df.values[:, 1]
x3 = df.values[:, 2]
y = df.values[:, 3]

display(df)
```

```python
m = len(y)
x0 = np.ones((m,1))
x1_1 = x1.reshape(m,1)
x1 = np.hstack((x0, x1_1))
x2_1 = x2.reshape(m, 1)
x2 = np.hstack((x0, x2_1))
x3_1 = x3.reshape(m, 1)
x3 = np.hstack((x0, x3_1))
```

```python
theta = np.zeros(2)
```

```python
def compute_cost(x, y, theta):

  prediction = x.dot(theta)
  error = np.subtract(prediction, y)
  sqrError = np.square(error)
  L = 1 / (2 * m) * np.sum(sqrError)

  return L
```

```python
cost = compute_cost(x1, y, theta)
print("Cost for theta_0 and theta_1 = ", cost)
```

```python
def grad_descent(x, y, theta, alpha, iterations):

  n = len(y)
  costhistory = np.zeros(iterations)

  for i in range(iterations):
    prediction = x.dot(theta)
    error = np.subtract(prediction, y)
    sum_delta = (alpha / n) * x.transpose().dot(error)
    theta = theta - sum_delta
    costhistory[i] = compute_cost(x, y, theta)

  return theta, costhistory
```

```python
theta1 = np.zeros(2)
theta2 = np.zeros(2)
theta3 = np.zeros(2)
iterations = 1500
alpha = .01
```

```python
theta1, costhistory1 = grad_descent(x1, y, theta1, alpha, iterations)
print('Final val of theta =',  theta1)
print('Cost history = ', costhistory1)
```

```python
plt.scatter(x1[:, 1], y, color = 'blue', marker = 'x', label = 'Training Data')
plt.plot(x1[:, 1], x1.dot(theta1), color = 'black', label = 'Linear Regression')
plt.grid(True)
plt.title('X1 Linear Regression Fit')
plt.xlabel('X1')
plt.ylabel('Y')
plt.legend()
```

```python
plt.plot(range(1, iterations + 1), costhistory1, color = 'purple')
plt.grid(True)
plt.title('Gradient Descent Convergence for X1')
plt.xlabel('Iterations')
plt.ylabel('Cost (L)')
```

```python
theta2, costhistory2 = grad_descent(x2, y, theta2, alpha, iterations)
print('Final val of theta =',  theta2)
print('Cost history = ', costhistory2)

plt.scatter(x2[:, 1], y, color = 'red', marker = 'x', label = 'Training Data')
plt.plot(x2[:, 1], x2.dot(theta2), color = 'black', label = 'Linear Regression')
plt.grid(True)
plt.title('X2 Linear Regression Fit')
plt.xlabel('X2')
plt.ylabel('Y')
plt.legend()
```

```python
plt.plot(range(1, iterations + 1), costhistory2, color = 'orange')
plt.grid(True)
plt.title('Gradient Descent Convergence for X2')
plt.xlabel('Iterations')
plt.ylabel('Cost (L)')
```

```python
theta3, costhistory3 = grad_descent(x3, y, theta3, alpha, iterations)
print('Final val of theta =',  theta3)
print('Cost history = ', costhistory3)

plt.scatter(x3[:, 1], y, color = 'green', marker = 'x', label = 'Training Data')
plt.plot(x3[:, 1], x3.dot(theta3), color = 'black', label = 'Linear Regression')
plt.grid(True)
plt.title('X3 Linear Regression Fit')
plt.xlabel('X3')
plt.ylabel('Y')
plt.legend()
```

```python
plt.plot(range(1, iterations + 1), costhistory3, color = 'cyan')
plt.grid(True)
plt.title('Gradient Descent Convergence for X3')
plt.xlabel('Iterations')
plt.ylabel('Cost (L)')
```
